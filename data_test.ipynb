```python
fileLocation='/home/neil/Desktop/1000.json'

import json
with open(fileLocation) as f:
    content = f.readlines()
# you may also want to remove whitespace characters like `\n` at the end of each line
content = [x.strip() for x in content] 
```

```python
textAnalyzers=[]
for eachItem in content:
    textAnalyzers.append(eval(eachItem)['title'][0]+eval(eachItem)['text'][0]+eval(eachItem)['cdcproduct'][0])
```
